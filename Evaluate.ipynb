```python
import Beta
import json
```

```python
with open('Data/labels.json', 'r', encoding='utf-8') as labelsfile:
    labels = json.load(labelsfile)
```

```python
max_num = 57
for i in range(1, max_num+1):
    filename = 'Data/%d.pdf'%i
    tmp = Beta.extract_from_pdf(filename)
    tag = '%d'%i
    result = labels.get(tag)
    print(filename)
    if tmp==result:
        print("Good job!")
    else:
        print("Wrong output:")
        print("output: ", tmp)
        print("label:  ", result)
```
