```python
def toLower(s):
    if type(s) == type(u""):
        return s.lower()
    return unicode(s, "utf8").lower().encode("utf8")

# def containTotalCostByWord(row):
#     regex = 'số tiền viết bằng chữ'
#     for cell in row:
#         found = re.search(regex, cell, re.IGNORECASE)
#         if found:
#             return True
    
def containTotalCostByWord(cell):
    digit_dict = ['không', 'một', 'mốt', 'hai', 'ba', 'bốn', 'tư', 'năm', 'lăm', 'sáu', 'bảy', 'tám', 'chín', 'mười']
    must_have = ['nghìn', 'đồng']

    for reg in must_have:
        found = re.search(reg, cell, re.IGNORECASE)
        if not found: 
            return False
        
    for reg in digit_dict:
        found = re.search(reg, cell, re.IGNORECASE)
        if found:
            return True
        
    return False

def preprocess(cell):
    result = []
    
    # remove unnecessary characters
    cell = cell.replace(',', '')
    cell = cell.replace('.', '')
    
    # convert to all lower case
    cell = toLower(cell)
    
    # split to list of words
    result = cell.split()
    
    return result

def getTotalCostByWord(cell):
    result = ''
    vector = preprocess(cell)
    begin = 0
    end = -1
    
    all_dict = ['không', 'một', 'mốt', 'hai', 'ba', 'bốn', 'tư', 'năm', 'lăm', 'sáu', 'bảy', 'tám', 'chín', 'mười', \
                'mươi', 'trăm', 'nghìn', 'triệu', 'tỷ', 'linh', 'lẻ']
    
    while True:
        if end<len(vector)-1:
            vector = vector[end+1:]
        else:
            break
            
        if 'đồng' in vector:
            end = vector.index('đồng')
            result = vector[end]
            for begin in reversed(range(end)):
                if vector[begin] in all_dict:
                    result = vector[begin] + ' ' + result
                else:
                    if checkTotalCostByWord(result):
                        return result
                    else:
                        break
        else:
            break
            
    return result

def preprocess(text):
    result = []
    
    # remove unnecessary characters
    text = text.replace(',', '')
    text = text.replace('.', '')
    
    # convert to all lower case
    text = toLower(text)
    
    # split to list of words
    result = text.split()
    
    return result

def checkTotalCostByWord(text):
    vector = preprocess(text)
    
    digit_dict = {'không': 0, 'một': 1, 'mốt': 1, 'hai': 2, 'ba': 3, 'bốn': 4, 'tư': 4, 'năm': 5, 'lăm': 5, 'sáu': 6, 'bảy': 7, 'tám': 8, 'chín': 9, \
                  'mười': 10, 'mười một': 11, 'mười hai': 12, 'mười ba': 13, 'mười bốn': 14, 'mười lăm': 15, 'mười sáu': 16, 'mười bảy': 17, 'mười tám': 18, 'mười chín': 19}
    post_dict = {'tỷ': 1000000000, 'triệu': 1000000, 'nghìn': 1000, 'trăm': 100, 'mươi': 10, 'đồng': 1}
    other_dict = {'linh': 0, 'lẻ': 0}
    i = 0
    digit = 0
    post = 0
    
    while i<len(vector):
        if i<len(vector)-1 and (vector[i] + ' ' + vector[i+1]) in digit_dict:
            if digit:
                return False
            digit = 1
            post = 0
            i = i+1
        elif vector[i] in digit_dict:
            if digit:
                return False
            digit = 1
            post = 0
        elif i>0 and vector[i] in post_dict:
            if i==len(vector)-1:
                if vector[i]=='đồng':
                    return True
                else:
                    return False
            if post>post_dict.get(vector[i]) or (post==post_dict.get(vector[i]) and post!='tỷ'):
                return False
            digit = 0
            post = post_dict.get(vector[i])
        elif i>0 and vector[i] in other_dict:
            if digit or (i==len(vector)-1) or (vector[i+1] not in digit_dict):
                return False
        else:
            return False            
        i = i+1
    
    return True
            
def convertWordToCost(text):
    result = 0
    vector = preprocess(text)
    
    n = len(vector)
    
    post_dict = {'tỷ': 1000000000, 'triệu': 1000000, 'nghìn': 1000, 'ngàn': 1000, 'trăm': 100, 'mươi': 10, 'đồng': 1}
    digit_dict = {'không': 0, 'một': 1, 'mốt': 1, 'hai': 2, 'ba': 3, 'bốn': 4, 'tư': 4, 'năm': 5, 'nhăm': 5, 'lăm': 5, \
                  'sáu': 6, 'bảy': 7, 'tám': 8, 'chín': 9, 'mười':10}
    
    stack = []
    
    l = 0
    
    for i in range(n):
        word = vector[i]
        if word in post_dict:
            val = post_dict.get(word)
            sum = 0
            while (stack) and (stack[-1] < val):
                sum += stack.pop()
            stack.append(sum*val)
        elif word in digit_dict:
            val = digit_dict.get(word)
            stack.append(digit_dict.get(word))          

    while stack:
        result += stack.pop()
    return result

def costToNumString(cost):
    result = ''
    count = 0
    while cost>0:
        digit = cost % 10
        if count == 3:
            result = '.' + result
            count = 0
        result = str(digit) + result
        count = count+1
        cost = cost // 10
    return result

#get totalCostWithVAT    
def getTotalCostWithVAT(tables):
    result = ''
    
    for i in range(len(tables)):
        table = tables[i].df
        m, n = table.shape
        for i in range(m):
            row = table.iloc[i,:].values
            for j in range(n):
                cell = row[j]
                if containTotalCostByWord(cell):
                    result = getTotalCostByWord(cell)
                    return costToNumString(convertWordToCost(result))
    return result
```
